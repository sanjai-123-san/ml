```python
from sklearn.impute import SimpleImputer
```

```python
## numerical processing pipeline
import numpy as np
numeric_processor=Pipeline(steps=[("imputation_mean",SimpleImputer(missing_values=np.nan,strategy="mean")),("scaler",StandardScaler())])
numeric_processor
```

```python
##categorical procesing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor=Pipeline(steps=[("imputation_consatnt",SimpleImputer(fill_value="missing",strategy="constant")),("onehot",OneHotEncoder(handle_unknown="ignore"))])
categorical_processor
```

```python
## combine processing technqiues
from sklearn.compose import ColumnTransformer
preprocessor=ColumnTransformer([("categorical",categorical_processor,["gender","City"]),("numerical",numeric_processor,["age","height"])])
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(preprocessor,LogisticRegression())
pipe
```
