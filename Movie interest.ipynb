```python
import pandas as pd
movie_info=pd.read_csv("movie Interests.csv")
```

```python
movie_info
```

```python
input_datasets=movie_info.drop(columns=['Interest'])
```

```python
input_datasets
```

```python
output_datasets=movie_info.drop(columns=['Interest'])
```

```python
output_datasets
```

```python
output_dataset=movie_info['Interest']
```

```python
from sklearn.tree import DecisionTreeClassifier
movie_model=DecisionTreeClassifier()
movie_model.fit(input_datasets,output_datasets)
movie_interest=movie_model.predict([[9,1],[33,0]])
```
